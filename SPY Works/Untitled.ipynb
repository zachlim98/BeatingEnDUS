```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import plotly.express as px
from statistics import NormalDist
```

```python
SPY = yf.download("SPY")
```

```python
SPY['Adj Close'] = SPY['Adj Close'].pct_change()*100
```

```python
SPY.reset_index(inplace=True)
SPY['Date_str'] = SPY["Date"].astype(str)
```

```python
SPY
```

```python
SPY_violin = px.violin(SPY.reset_index(), x = "Adj Close", orientation="h", 
labels={"variable":" ",
"value":"Percentage Change"},
title="Daily Percentage Change of SPY", hover_data=['Adj Close', 'Date_str'])

SPY_violin.show()
```

```python
SPY_hist = px.histogram(SPY.pct_change()*100, x = "Adj Close",
labels={"variable":" ",
"value":"Percentage Change"},
title="Daily Percentage Change of SPY")

SPY_hist.show()
```
